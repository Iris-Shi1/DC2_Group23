```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from tqdm import tqdm
```

```python
crime_df = pd.read_csv(r"C:\Users\assis\Downloads\filtered_crime_data.csv")
crime_df
```

```python
imp_crime_df = crime_df[['Crime ID', 'Month', 'Crime type', 'Borough']]
imp_crime_df
```

```python
pivot_df = imp_crime_df.pivot_table(index=['Month', 'Borough'], 
                                    columns='Crime type', 
                                    aggfunc='size', 
                                    fill_value=0).reset_index()

pivot_df
```

```python
month_data = [
    {'Month_Number': 49, 'Month_Year': 'Apr/15', 'MONTH': '49 (Apr 2015)'},
    {'Month_Number': 50, 'Month_Year': 'May/15', 'MONTH': '50 (May 2015)'},
    {'Month_Number': 51, 'Month_Year': 'Jun/15', 'MONTH': '51 (Jun 2015)'},
]

import calendar
from datetime import datetime

start_date = datetime(2010, 12, 1)
end_date = datetime(2024, 1, 1)
current_date = start_date

while current_date <= end_date:
    month_year = current_date.strftime('%b/%y')
    month_number = (current_date.year - 2015) * 12 + current_date.month + 45
    month_name = calendar.month_name[current_date.month]
    month_data.append({
        'Month_Number': month_number,
        'Month_Year': month_year,
        'MONTH': f'{month_number} ({month_name} {current_date.year})'
    })
    next_month = current_date.month % 12 + 1
    next_year = current_date.year + (current_date.month // 12)
    current_date = datetime(next_year, next_month, 1)

month_df = pd.DataFrame(month_data)

month_df
```

```python
merged_df = pd.merge(pivot_df, month_df[['Month_Number', 'Month_Year']], on='Month_Year', how='left')
merged_df
```

```python
merged_df.to_csv('crime_type_data.csv', index=False)
```

```python
all_df = pd.read_csv(r"C:\Users\assis\Downloads\PAS+crime+extra_final.csv")
all_df
```

```python
new_df = all_df.copy()
new_df['Q60'] = new_df['Q60'].replace({1: 1, 2: 1, 3: 2, 4: 3, 5: 3})
new_df
```

```python
top_20_conf = ['WT_Q63_B', 'SQ79DB', 'NQ135BDE', 'WT_Q58', 'XQ122A', 'NQCV28', 'Violence and sexual offences', 'XQ128A2B', 'WT_Q62', 'WT_Q64_B', 'NNQ27E', 'Q141', 'Borough', 'Vehicle crime', 'WT_Q44_E', 'NQCV7', 'Q62TI', 'WT_Q45_B', 'SQ132K', 'NQCV1', 'Q60']
top_20_trust = ['WT_Discrete', 'selfemployed_number', 'XQ122B', 'WT_Q45_B', 'WT_Q59', 'Vehicle crime', 'NQ57AA', 'White','Q62TG', 'WT_Q50_B', 'SQ109CAA', 'Monthly_pay_parttime', 'Monthly_pay_fulltime', 'Q1', 'WT_Q44_E', 'SQ132O', 'Q79H', 'WT_Q62', 'Q10F', 'PQ135AS', 'Q60']
```

```python
confidence_df = new_df[top_20_conf]
trust_df = new_df[top_20_trust]
trust_df
```

```python
trust_df = trust_df.dropna(subset=['Q60'])
trust_df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer

X = trust_df.drop(columns=['Q60'])
y = trust_df['Q60']
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Impute missing values using Iterative Imputer
imputer = IterativeImputer()
X_imputed = imputer.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
# Create LightGBM dataset
train_data = lgb.Dataset(X_train, label=y_train)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# Set parameters
params = {
    'objective': 'multiclass',
    'num_class': 4,  # Update based on the number of categories
    'metric': 'multi_logloss',
    'boosting_type': 'gbdt',
    'learning_rate': 0.01,
    'num_leaves': 31,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': -1
}

# Train the model
gbm = lgb.train(params,
                train_data,
                num_boost_round=1000,
                valid_sets=[train_data, test_data],
                early_stopping_rounds=100)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
# Predict on test set
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
y_pred_max = np.argmax(y_pred, axis=1)

print('Accuracy:', accuracy_score(y_test, y_pred_max))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_max)
print('Confusion Matrix:')
print(cm)

# Classification report
cr = classification_report(y_test, y_pred_max)
print('Classification Report:')
print(cr)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

y_train = y_train.astype(int)
y_test = y_test.astype(int)
y_train_encoded = y_train - 1
y_test_encoded = y_test - 1

classes = [1, 2, 3, 4, 5]

# Create and train the XGBoost model
xgb_model = XGBClassifier(eval_metric='mlogloss', use_label_encoder=True, classes=classes)
xgb_model.fit(X_train, y_train_encoded)
y_pred = xgb_model.predict(X_test)

print('Accuracy:', accuracy_score(y_test_encoded, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test_encoded, y_pred))
print('Classification Report:')
print(classification_report(y_test_encoded, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from tqdm import tqdm

# Create and train the Random Forest model with a progress bar
rf_model = RandomForestClassifier(n_estimators=100)  # You can adjust the number of estimators
for _ in tqdm(range(100), desc='Training', unit='trees'):
    rf_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf_model.predict(X_test)

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
