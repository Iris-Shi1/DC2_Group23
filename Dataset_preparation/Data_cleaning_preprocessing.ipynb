```python
add_data['Weekly_pay_parttime'] = pd.to_numeric(add_data['Weekly_pay_parttime'], errors='coerce')
add_data['Monthly_pay_parttime'] = pd.to_numeric(add_data['Monthly_pay_parttime'], errors='coerce')
```

```python
def clean_and_convert(column):
    # Remove commas
    column = column.str.replace(',', '', regex=False)
    # Convert to numeric with coercion for non-convertible values
    column = pd.to_numeric(column, errors='coerce')
    return column


columns_to_convert = [
    'White', 'Asian', 'Black', 'Mixed/ Other', 'Total', 
    'Employees_number', 'selfemployed_number', 
    'fulltime_work_number', 'parttime_work_number'
]

# Apply the function to the specified columns
for column in columns_to_convert:
    add_data[column] = clean_and_convert(add_data[column])

# Convert to integer (Int64 to handle possible NaNs)
for column in columns_to_convert:
    add_data[column] = add_data[column].astype('Int64')
```
