```python
import pandas as pd

df = pd.read_csv('../project_data/combined_crime_outcomes.csv', low_memory=False)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Check unique values in the 'Month' column
valid_dates = df['Month'][df['Month'] != 'Month']
df['year'] = pd.to_datetime(valid_dates).dt.year
df['month_num'] = pd.to_datetime(valid_dates).dt.month

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
top_outcome_types = df['Outcome type'].value_counts().head(10).index.tolist()

# Filter the DataFrame for the desired outcome types
filtered_df = df[df['Outcome type'].isin(top_outcome_types[:2])]


# Step 3: Group and Aggregate
grouped = filtered_df.groupby(['year', 'month_num', 'Outcome type']).size().unstack(fill_value=0)

# Combine year and month_num to form a date column
grouped['date'] = grouped.index.map(lambda x: pd.Timestamp(year=int(x[0]), month=int(x[1]), day=1))

# Step 4: Plotting
plt.figure(figsize=(12, 8))

for outcome in top_outcome_types[:2]:
    plt.plot(grouped['date'], grouped[outcome], label=outcome)

plt.title('Frequency Distribution of Common Outcome Types by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../project_data/combined_crime.csv')
print(df.columns)


valid_dates = df['Month'][df['Month'] != 'Month']
df['year'] = pd.to_datetime(valid_dates).dt.year
df['month_num'] = pd.to_datetime(valid_dates).dt.month

import numpy as np

top_crime_types = df['Crime type'].value_counts().head(10).index.tolist()
filtered_df = df[df['Crime type'].isin(top_crime_types)]
filtered_df = filtered_df[filtered_df['year'].isin([2016,2017,2018,2019,2020,2021,2022,2023])]

# Step 3: Group and Aggregate
grouped = filtered_df.groupby(['year', 'month_num', 'Crime type']).size().unstack(fill_value=0)

# Combine year and month_num to form a date column
grouped['date'] = grouped.index.map(lambda x: pd.Timestamp(year=int(x[0]), month=int(x[1]), day=1))
fig, ax = plt.subplots(figsize=(12, 8))


for outcome in top_crime_types:
    plt.plot(grouped['date'], grouped[outcome], label=outcome)

# Manually set x-axis ticks at monthly intervals
date_range = pd.date_range(start=min(grouped['date']), end=max(grouped['date']), freq='MS')
plt.xticks(date_range, [d.strftime('%Y-%m') for d in date_range], rotation=90)

plt.title('Frequency Distribution of Crime Types by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Frequency')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../project_data/combined_crime_stop_and_search.csv')


print(df.columns)
for column in df.columns:
    print(repr(column))


valid_dates = df['Date'][df['Date'] != 'Date']
df['year'] = pd.to_datetime(valid_dates).dt.year
df['month_num'] = pd.to_datetime(valid_dates).dt.month


top = df['Age range'].value_counts().head(10).index.tolist()
filtered_df = df[df['Age range'].isin(top)]

# Step 3: Group and Aggregate
grouped = filtered_df.groupby(['year', 'month_num', 'Age range']).size().unstack(fill_value=0)

# Combine year and month_num to form a date column
grouped['date'] = grouped.index.map(lambda x: pd.Timestamp(year=int(x[0]), month=int(x[1]), day=1))
# Step 4: Plotting
plt.figure(figsize=(12, 8))

for type in top:
    plt.plot(grouped['date'], grouped[type], label=type)

plt.title('Frequency Distribution of Stop-and-Search Age Ranges by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

df = pd.read_csv('project_data/combined_crime.csv')

# Step 1: Get the top 15 most common values of the 'Location' column
top_locations = df['Location'].value_counts().head(15).index.tolist()
top_crime_types = df['Crime type'].value_counts().head(10).index.tolist()

# Step 2: Create a subset of the DataFrame containing only the top 15 locations and the top 10 Outcome types
filtered_df = df[df['Location'].isin(top_locations) & df['Crime type'].isin(top_crime_types)]


# Step 3: Create a pivot table to calculate the frequency of each Outcome type for each Location
pivot_table = pd.pivot_table(filtered_df, index='Location', columns='Crime type', aggfunc='size', fill_value=0)


# Step 4: Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Frequency of Crime Types by Location')
plt.xlabel('Crime Type')
plt.ylabel('Location')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the crime data
# Assuming 'Month' column contains datetime information


valid_dates = df['Month'][df['Month'] != 'Month']
df['year'] = pd.to_datetime(valid_dates).dt.year
df['month_num'] = pd.to_datetime(valid_dates).dt.month

# Group by year and count the number of crimes
crime_counts_by_year = df.groupby('year').size()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(crime_counts_by_year.index, crime_counts_by_year.values, marker='o', linestyle='-')
plt.title('Number of Crimes Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.xticks(crime_counts_by_year.index, rotation=45)
plt.tight_layout()
plt.show()
```
